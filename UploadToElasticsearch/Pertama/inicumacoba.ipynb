```python
hasil = []
for index, row in df.iterrows():

    try:
        tweet = TextBlob(row['Mentions'])
        # output sentiment polarity
        print (tweet.sentiment.polarity)
        # determine if sentiment is positive, negative, or neutral
        if tweet.sentiment.polarity < 0:
            sentiment = "negative"
        elif tweet.sentiment.polarity == 0:
            sentiment = "neutral"
        else:
            sentiment = "positive"

        message = row['Mentions']
    except:
        sentiment = "null"
        message = "null"
    
    hasil.append(sentiment)


```

```python
df['Sentiment'] = hasil
df.to_csv("hasil-out.csv", index=False, encoding = "ISO-8859-1")
```
